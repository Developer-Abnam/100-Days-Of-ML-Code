# **Analyzing Selling Price of used Cars using Python**

**Import Required Python Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

**Load the Dataset**

```python
df = pd.read_csv('/content/selling_car.csv')
df.shape
```

**Assign Column Headers**

```python
headers = ["symboling", "normalized-losses", "make",
           "fuel-type", "aspiration","num-of-doors",
           "body-style","drive-wheels", "engine-location",
           "wheel-base","length", "width","height", "curb-weight",
           "engine-type","num-of-cylinders", "engine-size",
           "fuel-system","bore","stroke", "compression-ratio",
           "horsepower", "peak-rpm","city-mpg","highway-mpg","price"]

df.columns = headers
df.head()
```

**Check for Missing Values**

```python
data = df
data.isna().any()
```

**Convert MPG to L/100km**

```python
data['city-mpg'] = 235 / data['city-mpg']
data.rename(columns = {'city_mpg': "city-L / 100km"}, inplace=True)
print(data.columns)
data.dtypes
```

**Convert Price Column to Integer**

```python
data.price.unique()
data = data[data.price != '?']
data.price = data.price.astype(int)
data.dtypes
```

**Normalize Features**

```python
data['length'] = data['length'] / data['length'].max()
data['width'] = data['width'] / data['width'].max()
data['height'] = data['height'] / data['height'].max()

bins = np.linspace(min(data['price']), max(data['price']), 4)
group_names = ['low', 'medium', 'high']
data['price_binned'] = pd.cut(data['price'], bins, labels=group_names, include_lowest=True)
print(data['price_binned'])
plt.hist(data['price_binned'])
plt.show()
```

**Convert Categorical Data to Numerical**

```python
pd.get_dummies(data['fuel-type']).head()
data.describe()
```

**Data Visualization**

```python
plt.boxplot(data['price'])
sns.boxplot(x=data['drive-wheels'], y=data['price'])
sns.scatterplot(x=data['engine-size'], y=data['price'])
plt.title('Scatterplots of Engine size Vs Price')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.grid()
plt.show()
```

**Grouping Data by Drive-Wheels and Body-Style**

```python
test = data[['drive-wheels', 'body-style', 'price']]
data_grp = test.groupby(['drive-wheels', 'body-style'], as_index=False).mean()
data_grp
```

```python
data_pivot = data_grp.pivot(index = 'drive-wheels',
                            columns = 'body-style')
data_pivot

plt.pcolor(data_pivot, cmap ='RdBu')
plt.colorbar()
plt.show()
```

```python
data_annova = data[['make', 'price']]
grouped_annova = data_annova.groupby(['make'])
annova_results_l = sp.stats.f_oneway(
                             grouped_annova.get_group('honda')['price'],
                             grouped_annova.get_group('subaru')['price']
                                    )
print(annova_results_l)

sns.regplot(x ='engine-size', y ='price', data = data)
plt.ylim(0, )
```
